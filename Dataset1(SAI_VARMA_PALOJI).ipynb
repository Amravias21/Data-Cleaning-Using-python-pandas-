#Age

```python
df.age.isnull().sum()
```

```python
df.age.head()
```

```python
df.age.max()
```

```python
df.age.min()
```

```python
median_age = df.age.median()
median_age
```

```python
df.age = df.age.fillna(median_age)
df.age
```

```python
df.age.isnull().sum()
```

```python
df.age = df.age.astype(int)
df.age
```

Established Year


```python
df.established.head()
```

```python
df.established = df.established.replace( -1, np.nan)
df.established = df.established.fillna("Unknown")
df.established
```

```python
df.head()
```

#Salary

```python
df.salary.head(10)
```

```python
df.salary = df.salary.apply(lambda x:x.replace('$',''))
df.salary = df.salary.apply(lambda x:x.replace('k','000'))
df.salary.head()
```

```python
df[['min-salary', 'max-salary']] = df.salary.str.split("-", expand = True)
df.pop('salary')
```

```python
df.head()
```

```python
df['min-salary'] = df['min-salary'].astype(int)
df['max-salary'] = df['max-salary'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

#Rating

```python
df.rating.head(10)
```

```python
df.rating.isnull().sum()
```

```python
avg_rating = df.rating.mean()
avg_rating
```

```python
df.rating = df.rating.fillna(avg_rating)
df.rating
```

```python
df['rating'] = df['rating'].apply(lambda x: x if x>=0 else avg_rating)
df.rating = df.rating.round(decimals=1)
df.rating
```

#Easy Apply

```python
df['easy-apply'] = df['easy-apply'].apply(lambda x: 1 if x=="TRUE" else 0).astype(bool)
```

```python
df['easy-apply'].head()
```

#Location


```python
df['location'] = df['location'].str.replace('Australia Aus', 'Australia,Aus')
df['location'] = df['location'].str.replace('India In', 'India,In')
```

```python
df[['country', 'country-code']] = df.location.str.split(",", expand = True)
```
